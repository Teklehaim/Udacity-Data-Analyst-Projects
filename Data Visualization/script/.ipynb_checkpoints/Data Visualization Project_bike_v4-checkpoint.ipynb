### Reading files

```python
# reading the file 

def reading_file(path, filename):
    ''' This function reads the file and returns a pandas DataFrame of it'''

    with open(os.path.join(path, filename), "r") as file_object:

        df = pd.read_csv(file_object, sep=",", header = 0)

        return df
    
#Function parameters:

path_raw_data = r"..\data\raw_data"                    # relative Path where the file exists 
file_bike = "201902-fordgobike-tripdata.csv"           # File name to be read

# Calling the function:

bike_original = reading_file(path = path_raw_data, filename = file_bike)

bike_df = bike_original.copy()
```

##  Assessing data 

### Quality issues 

```python
# Checking for data type:
bike_df.info()
```

As we can see in the above, some of the data types are not appropriate. Therefore, we need to convert them into more appropriate datatype.

The following quality issues are observed in this dataset


- Missing values are observed in the member_birth_year, station_ids, station_names and member_gender variables
- The data type of user_type,bike_share_for_all_trip and member_gender are not appropriate (they should be changed into categorical data type)ogrical
- The dates are not in an appropriate data type ( normally, they should be in datetime) 
- The start_station_id and start_station_id are float data type ( normally, they supposed be integers)

## Cleaning 

### Define


- Drop Missing values in the member_birth_year,member_gender,station_ids and station_names variables
- Convert the data type of user_type, bike_share_for_all_trip and member_gender into categorical data type
- Convert the start_time and member_birth_year into datetime data type 
- Convert the data type of start_station_id and start_station_id into integer data type

### Code

```python
# Drop missing values
missing_data_columns = (["member_gender", "member_birth_year", "start_station_id",
                         "end_station_id", "start_station_name", "end_station_name"])

bike_df.dropna(subset=missing_data_columns, axis=0, inplace=True)
```

```python
# Convert the user_type and member_gender into categorical data type:
columns = ["member_gender", "user_type", "bike_share_for_all_trip"]

for column in columns:

    bike_df[column] = bike_df[column].astype("category")
```

```python
# Convert the user_type and member_gender into categorical data type:
columns = ["member_gender", "user_type", "bike_share_for_all_trip"]

for column in columns:

    bike_df[column] = bike_df[column].astype("category")
```

```python
# Convert the start_time and end_time into datetime and member_birth_year into integer type

bike_df["member_birth_year"] = bike_df["member_birth_year"].fillna(
    0).apply(lambda x: int(x))

dates_columns = ["start_time", "end_time"]

for date_column in dates_columns:

    bike_df[date_column] = pd.to_datetime(
        bike_df[date_column], format="%Y-%m-%d %H:%M:%S")
```

```python
# Convert the data type of start_station_id and end_station_id into integer data type

column_start_ids = ["start_station_id", "end_station_id"]

for column in column_start_ids:

    bike_df[column] = bike_df[column].apply(lambda x: int(x))
```

```python
# Drop the row containing age of 141. This must be dropped.

bike_df.drop(labels=27370, inplace=True, axis=0)
```

## Exploratory 

### Univariante Exploratory # 1:

Which gender group tend  to share bike in all trips?

```python
bike_df[bike_df["bike_share_for_all_trip"].isin(["Yes"])]["member_gender"].value_counts(
)/bike_df[bike_df["bike_share_for_all_trip"].isin(["Yes"])].shape[0]*100
```

```python
fig, ax = plt.subplots(figsize = (16,9))

(bike_df[bike_df["bike_share_for_all_trip"].isin(["Yes"])]["member_gender"].value_counts(
)/bike_df[bike_df["bike_share_for_all_trip"].isin(["Yes"])].shape[0]*100).plot(kind="bar");

plt.title("Bike sharing in all trips based on gender", fontsize = 20);
plt.xlabel("Gender", fontsize = 16);
plt.ylabel("Number of bike user [%]", fontsize = 16);
plt.grid()
plt.show();
```

**Observation** : Mainly males share bikes which accounts for 75% followed by female with 20%.

### Explanatory #1 :

#### Which user and gender categories use bike the most?

```python
sb.catplot(data=bike_df, x="member_gender", hue="user_type", kind="count",palette=sb.color_palette(['cyan', 'orange']), height = 8, aspect = 8/6)
plt.xlabel(" Gender", fontsize=16)
plt.ylabel(" Numbers of customers", fontsize=16)

plt.title(" The number of user and gender categories using bike", fontsize=20)
#plt.legend(loc="best", title="Users type", fontsize=15)
plt.xticks(fontsize=15);
plt.yticks(fontsize=15);
plt.grid()
plt.show();
```

**Observation** : Among the bike users, the subscribers are the majority who use ride bikes. Additionally, males riders account four times the number of females. 

### Explanatory # 2 :

#### During which time of the day do people use bike?

```python

sb.catplot(data=bike_df, x="starting_hour", kind="count", hue ="user_type",height = 12, aspect=12/8, legend = False,palette=sb.color_palette(['k', 'r']));

plt.xlabel("Biking starting hours", fontsize = 16);
plt.ylabel("Number of bikers in each hour", fontsize = 16);
plt.title("The hours in which people start biking", fontsize = 20);
plt.legend(loc ="best", title ="User type", fontsize = 16)
plt.xticks(fontsize = 15);
plt.yticks(fontsize = 15);
plt.grid()
plt.show()
```

**Observation** : From the bar plots above, we observe that most people start biking in the early hours 7 to 9 am and in the after noon 4 to 7 pm. In addition to this, can see that less bikers and relatively the same number of bikers from 10 am to 3 pm.
Apart from this, it is seen that subscribers constitute the highest number of bikers compared to the customers.

### Explanatory # 3 :

#### In which day of the week do people bike more ?

```python
order_days = ["Saturday", "Sunday", "Monday",
              "Friday", "Wednesday", "Tuesday", "Thursday", ]

sb.catplot(data=bike_df, x="starting_day",
           hue="member_gender", kind="count", order=order_days, height = 11, aspect = 11/9, legend = False)
plt.xticks(rotation=45);

plt.xlabel(" Days of the week", fontsize=16);
plt.ylabel(" Number of users", fontsize=16);

plt.title(" Number of bike users on a daily basis.", fontsize=20);
plt.xticks(fontsize=15);
plt.yticks(fontsize=15);
plt.legend(loc = "best", title ="Gender", fontsize = 15 )
plt.grid()
plt.show()
```

**Observation** : As we can see from the bar plot above, most people rides during the working days with the highest on Thurdays.
The number of riders during the weekends is less by half comparing during the week days.

### Explanatory # 4 :

#### Which gender group and user spends more time biking?

```python

bike_df["duration_min"] = bike_df["duration_sec"].apply(lambda x: x/60)

fig, ax = plt.subplots(figsize=(18, 8))

sb.barplot(data=bike_df, x="member_gender", y="duration_min", hue="user_type",palette=sb.color_palette(['green', 'cyan']));

plt.title("Time spend biking by different users and gender", fontsize=20);
plt.xlabel(" Gender", fontsize=16);
plt.ylabel(" Duration [min]", fontsize=16);

plt.legend(loc="best", title="Users type", fontsize=16);
plt.xticks(fontsize=15);
plt.yticks(fontsize=15);
plt.grid()
plt.show()
```

**Observation** : Among all user types and gender, the gender which is not identified (other) spends more time biking. Additionally, customers spend more time compared to the subscribers, nevertheless the maximum time spend is less than 30 minutes.

### Explanatory # 5 :

#### Is ther correlation among biker's age, time spend biking and starting hours?

```python
average_displacement = bike_df["displacement"].mean()

fig, ax = plt.subplots(figsize=(18, 8))

plt.scatter(data=bike_df, x="member_age", y="displacement", c="duration_min", cmap="plasma");

plt.axhline( average_displacement, color = "r")
plt.colorbar(label="Duration [min]");
plt.xlabel("Age [year]", fontsize=16);
plt.ylabel("Displacement [km]", fontsize=16);
plt.title("Correlation among bikers' age, time spend and displacement", fontsize = 20)

plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.show()
```

**Observation** : Most bikers are in the age ranging from 18 to 70 years of age. And the displacement travelled is often less than 10 km. Additionally, we see that  the number of people driving bike reduces as they are aged. The average displacement which is denoted by the horizontal line is about 1.89 km. 
Regarding the time spend, we see that most of the bikers spend not more than 3-4 hours.
